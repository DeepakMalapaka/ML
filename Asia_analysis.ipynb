# **Asia cup analysis**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
from plotly.offline import iplot
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data=pd.read_csv('/content/asiacup.csv')
data
```

```python
data.describe()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Team', hue='Result', data=data)
plt.title('Win/Loss Distribution by Team')
plt.show()
```

# **Observation:**
The image is a histogram titled "Distribution of Matches Played Over the Years", showing the number of matches played during specific time intervals.

X-axis: Represents the years, ranging from 1985 to 2020.

Y-axis: Represents the number of matches played, with values up to 25.

Bars: Blue bars indicate the number of matches played in each time interval.




```python
plt.figure(figsize=(12, 6))
sns.histplot(data['Year'], bins=20, kde=True)
plt.title('Distribution of Matches Played Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Matches')
plt.show()
```

# **Observation:**
The image is a histogram titled "Distribution of Matches Played Over the Years", showing the number of matches played during specific time intervals.

X-axis: Represents the years, ranging from 1985 to 2020.

```python
plt.figure(figsize=(12, 6))
boundaries = data.groupby('Team')[['Fours', 'Sixes']].sum()
boundaries.plot(kind='bar', stacked=False)
plt.title('Fours and Sixes by Team')
plt.show()
```

# **Observation:**
The image is a bar chart titled "Fours and Sixes by Team", comparing the number of fours and sixes hit by cricket teams.

Teams: The x-axis lists teams: Afghanistan, Bangladesh, Hong Kong, India, Pakistan, Sri Lanka, and UAE.

Runs (Fours and Sixes): The y-axis represents the count, going up to 1000.

Bars:

Blue bars represent fours.

Orange bars represent sixes.
